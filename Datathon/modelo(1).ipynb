```python
import joblib
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Ésta es una segunda prueba, aunque el modelo tuvo una performance inadecuada(inferior) al anterior, el próximo modelo se evaluará con otras métricas ya que aumentando el número de variables no se consigue mejorar el acurracy.

```python
prueba_modelo = pd.read_excel('E-Commerce_test.xlsx')
prueba_modelo1 =prueba_modelo[['Cost_of_the_Product','Weight_in_gms','Mode_of_Shipment','Product_importance','Customer_rating','Customer_care_calls']]
prueba_modelo1
```

```python
lbl_Encode = LabelEncoder()
prueba_modelo1['Mode_of_Shipment']= lbl_Encode.fit_transform(prueba_modelo1['Mode_of_Shipment'])
prueba_modelo1['Product_importance']= lbl_Encode.fit_transform(prueba_modelo1['Product_importance'])
```

```python
modelo1 = joblib.load('modelo_de_predicción(1).joblib')
prediccion = modelo1.predict(prueba_modelo1)
len(prediccion)
pd.DataFrame(prediccion).to_csv('Onjuan25.csv',index=False, header=['pred'])
```
